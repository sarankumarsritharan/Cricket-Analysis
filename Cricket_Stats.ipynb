```python
df = df.drop_duplicates()
```

```python
# Split up span into Start and End Date
df['Span'].str.split(pat = '-')
```

```python
df['Start_Year'] =  df['Span'].str.split(pat = '-').str[0]
```

```python
df['End_Year'] =  df['Span'].str.split(pat = '-').str[1]
```

```python
df = df.drop(['Span'],axis=1)
```

```python
df['Country']=df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country']=df['Country'].str.split(pat = ')').str[0]
```

```python
df['Country']
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat = '*').str[0]
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype(int)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.astype({'Start_Year':'int','End_Year':'int'})
```

```python
df.dtypes
```

```python
df['Matched'] = df['Matched'].astype(str).str.split(pat='*').str[0]
```

```python
df['Matched'] = df['Matched'].astype(int)
```

```python
df.dtypes
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype(str).str.split(pat='+').str[0]
df['Balls_Faced'] = pd.to_numeric(df['Balls_Faced'], errors='coerce').fillna(0).astype(int)
```

```python
df
```

```python
df = df.drop(58,axis=0)
```

```python
df
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].str.split(pat = '*').str[0]
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype(float)
```

```python
df
```

```python
df.dtypes
```

```python
# Find the Career length
df['Career_length'] = df['End_Year'] - df['Start_Year']
```

```python
df
```

```python
# What is the average Career length
df['Career_length'].mean()
```

```python
# Average Batting strike for cricketer who played over 10years
```

```python
df[df['Career_length']>10]['Batting_Strike_Rate'].mean()
```

```python
# find no of cricketer played before 1960
df[df['End_Year']<1960]['Player'].count()
```

```python
# Max Highest inns score by country
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highinscore').reset_index().sort_values(by='Highinscore',ascending=False)
```

```python
# Hundred,Fifties,Ducks avg by country
df[['100', '50', '0']] = df[['100', '50', '0']].apply(pd.to_numeric, errors='coerce')
df.groupby('Country')[['100', '50', '0']].mean()
```
